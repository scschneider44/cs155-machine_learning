```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
train_file = 'data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
#get alphas
alpha = np.linspace(0.01, 3, 30)
```

```python
weights = [[] for i in range(9)]
# fit the data with different alphas using lasso regression and append the weights to a list
for i in range(30):
    clf = Lasso(alpha[i])
    clf.fit(X_train, y_train)
    w = clf.coef_
    for j in range(9):
        weights[j].append(w[j])    
```

```python
plt.figure()
# plot the weight size vs alpha
for i in range(9):
    plt.plot(alpha, weights[i], linewidth = 1)
plt.legend(['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8', 'w9'], loc = 'right')
plt.xlabel('alpha')
plt.ylabel('size')
plt.title('Lasso')
plt.show()
```

```python
alpha = np.linspace(1, 10000, 10000)
weights = [[] for i in range(9)]
# fit the data using ridge regression with different alphas
for i in range(10000):
    clf = Ridge(alpha[i])
    clf.fit(X_train, y_train)
    w = clf.coef_
    for j in range(9):
        weights[j].append(w[j])  
```

```python
# plot the weight size vs regularization strength
plt.figure()
for i in range(9):
    plt.plot(alpha, weights[i], linewidth = 1)
plt.legend(['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8', 'w9'], loc = 'best')
plt.xlabel('alpha')
plt.ylabel('size')
plt.title('Ridge')
plt.show()
```
